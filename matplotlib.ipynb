```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.xlabel("day")
plt.ylabel("temperature")
plt.title("weather")
plt.plot(x,y,color="green",linewidth=5,linestyle="dotted") # there are many colours and line style google matplotlib official file
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
z=[55,52,45,30,20,29,40]
```

```python
plt.xlabel('day')
plt.ylabel('temp')
plt.title('weather')
plt.plot(x,y,label="Max")
plt.plot(x,z,label="min")
plt.legend()
```

```python
company=["google","amzn","wipro","tcs"]
revenue=[99,97,50,60]
profit=[50,20,30,30]
```

```python
xpos=np.arange(len(company))
xpos
```

```python
plt.xticks(xpos,company)
plt.xlabel("companies")
plt.ylabel("revenue")
plt.title("IT COMPANIES")
plt.bar(xpos-0.2,revenue,label="REVENUE",width=0.4)
plt.bar(xpos+0.2,profit,label="PROFIT",width=0.4)
plt.legend()
```

```python
plt.yticks(xpos,company)
plt.ylabel("companies")
plt.xlabel("revenue")
plt.title("IT COMPANIES")
plt.barh(xpos-0.2,revenue,label="REVENUE")
plt.barh(xpos+0.2,profit,label="PROFIT")
plt.legend()
```

```python
men_sugar=[80,90,100,240,150,95,98]
women_sugar=[100,90,90,80,95,150,120]
```

```python
plt.xlabel("sugar_range")
plt.ylabel("total_patients")
plt.title("patients_records")
plt.hist([men_sugar,women_sugar],bins=[80.,100,125,150],rwidth=0.95,label=['Men','Women'])
plt.legend()                                                                           
```

```python
plt.ylabel("sugar_range")
plt.xlabel("total_patients")
plt.title("patients_records")
plt.hist([men_sugar,women_sugar],bins=[80.,100,125,150],rwidth=0.95,label=['Men','Women'],orientation='horizontal')
plt.legend()                                                                           
```
