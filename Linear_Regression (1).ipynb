```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
#linear model: y=mx+b
df=pd.read_csv('homeprices.csv')
df
```

```python
df[['area']]
```

```python
df.price
```

```python
plt.scatter(df['area'],df['price'],color='red',marker='+')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
y=135.78767123*3300+180616.43835616432
y
```

```python
plt.xlabel("area")
plt.ylabel("price $")
plt.scatter(df['area'],df['price'],color='red',marker='+')
plt.plot(df.area,reg.predict(df[['area']]))
```

```python
df=pd.read_csv(r"C:\Users\suruk\Downloads\canada_per_capita_income.csv")
df
```

```python
df=df.rename(columns={'per capita income (US$)':'per_capita_income'})
df
```

```python
plt.xlabel("years")
plt.ylabel("per_capita_income(US$)")
plt.title("canada_per_capita_income")
plt.scatter(df['year'],df['per capita income'],color='red',marker='+')
```
